creating the sp object

```python
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id="170d500451dc4520965b237c4e9a28fc",client_secret=getpass.getpass(prompt='Password: ')))
```

```python
#function to get more than 100 songs

def get_playlist_tracks(username, playlist_id):
    
    results = sp.user_playlist_tracks(username, playlist_id)
    tracks = results['items']
    
    while results['next']:
        results = sp.next(results)
        tracks.extend(results['items'])
    
    return tracks

```

getting trackIDs for all the songs in the list of playlists

```python

for pl in playlists.keys():
    track_ids = []
    print(pl)
    tracks = get_playlist_tracks("spotify", playlists[pl])
    track_ids.extend([track['track']['id'] for track in tracks])
    time.sleep(1)
```

Removing duplicates

```python
from collections import OrderedDict

track_ids = list(OrderedDict.fromkeys(track_ids))
```

getting all song features 

```python
features = []
for x in range(int(len(track_ids) / 100)):
    features.extend(sp.audio_features(track_ids[(x*100):((x*100)+100)]))
features.extend(sp.audio_features(track_ids[(int(len(track_ids) / 100)*100):len(track_ids)]))
```
